# Setting up the environment

```python
import dotenv
import openai

dotenv.load_dotenv()
```

# First steps

```python
client = openai.OpenAI()
```

```python
user_message = "I have recently adopted a dog, can you suggest some dog names?"

completion = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are Frank, a chatbot that reluctantly answers questions with sarcastic responses."}, 
        {"role" : "user",
        "content" : user_message}])
```

```python
print(completion)
```

```python
print(completion.choices[0].message.content)
```

# Another exmaple : Sentiment analysis bot

# Exploring different parameters

```python
# Max tokens - Limit the number of tokens in the response

completion_4 = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are a chat bot that can answer questions and help with tasks."}, 
        {"role" : "user",
        "content" : "Can you briefly explain quantum physics to me?"}],
    max_tokens=250)

print(completion_4.choices[0].message.content)
```

```python
completion_5 = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are a chat bot that can answer questions and help with tasks."}, 
        {"role" : "user",
        "content" : "Can you briefly explain quantum physics to me?"}],
    max_tokens=50)

print(completion_5.choices[0].message.content)
```

```python
# Temperature - Control the randomness of the response

completion_6 = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are a chat bot that can answer questions and help with tasks."}, 
        {"role" : "user",
        "content" : "Can you briefly explain quantum physics to me?"}],
    max_tokens=250,
    temperature=0)

print(completion_6.choices[0].message.content)
```

```python
completion_7 = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are a chat bot that can answer questions and help with tasks."}, 
        {"role" : "user",
        "content" : "Can you briefly explain quantum physics to me?"}],
    max_tokens=250,
    temperature=0)

print(completion_7.choices[0].message.content)
```

```python
completion_8 = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are a chat bot that can answer questions and help with tasks."}, 
        {"role" : "user",
        "content" : "Can you briefly explain quantum physics to me?"}],
    max_tokens=250,
    temperature=2)

print(completion_8.choices[0].message.content)
```

```python
# Stream - Stream the response

completion_9 = client.chat.completions.create(
    model="gpt-4", 
    messages=[
        {"role" : "system", 
        "content" : "You are a chat bot that can answer questions and help with tasks."}, 
        {"role" : "user",
        "content" : "Can you briefly explain quantum physics to me?"}],
    max_tokens=250,
    temperature=0,
    stream=True)

for i in completion_9:
    print(i.choices[0].delta.content, end="")
```
